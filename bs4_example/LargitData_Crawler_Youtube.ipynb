```python
import requests
res = requests.get('https://www.youtube.com/watch?v=B692I8fSAo8', verify=False)
#print(res.text)
```

```python
import re
m = re.search('args":({.*?}),', res.text)
m.group(1)
```

```python
import json
jd = json.loads(m.group(1))
print(jd['url_encoded_fmt_stream_map'])
```

```python
import urlparse
a = urlparse.parse_qs(jd['url_encoded_fmt_stream_map'])
a
```

```python
print a['url'][0]
```

```python
import shutil

res2 = requests.get(a['url'][0], verify = False, stream = True)
f = open('a.mp4', 'wb')
shutil.copyfileobj(res2.raw, f)
f.close()
```
