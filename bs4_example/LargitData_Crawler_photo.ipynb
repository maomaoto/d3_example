```python
import requests
from bs4 import BeautifulSoup
head = {'User-Agent':"Mozilla/5.0 (Windows NT 6.1; WOW64; rv:43.0) Gecko/20100101 Firefox/43.0"}
res = requests.get('http://ck101.com/thread-3035889-1-1.html', headers = head)
print(res.text)
#soup = BeautifulSoup(res.text, 'lxml')
#for img in soup.select('.img'):
#    print img
```

```python
import shutil
soup = BeautifulSoup(res.text,'lxml')

for img in soup.select('img'):
    if img.get('file'):
        print(img.get('file'))
        s = img.get('file')
        fn = s.split('/')[-1]
        res2 = requests.get(s, stream = True)
        f = open(fn, 'wb')
        shutil.copyfileobj(res2.raw, f)
        f.close()
        del res2 #remove temporary data
```
